```python
def get_slogan(brand):
  templates=[
      f"{brand},让创意更精彩!",
      f"{brand},设计不止，创意不息。",
      f"选择{brand},看见更好的自己。",
      f"{brand},用心做好每一个作品。"
  ]
  return templates

print("===广告文案生成===")
name=input("请输入品牌/公司名：")

slogans=get_slogan(name)

print("\n为你生成的广告语")

for i in range(len(slogans)):
  print(f"{i+1}.{slogans[i]}")
```
